## Documentation 📝

### ❌ Common Documentation Issues:

```python
# Missing or unclear documentation
def clean(df):
    df = df.dropna()
    return df

# Vague comments
# Do some stuff to the data
def process_data(data):
    # Magic numbers with no explanation
    return data * 1.5 + 42
```

### ✅ Good Documentation:

```python
def clean_spotify_data(df: pd.DataFrame) -> pd.DataFrame:
    """Remove rows with missing values and standardize column names.
    
    Args:
        df: Raw DataFrame from Spotify API
        
    Returns:
        Cleaned DataFrame with standardized column names
    
    Example:
        >>> raw_df = pd.DataFrame({'Artist Name': ['Taylor', None]})
        >>> clean_spotify_data(raw_df)
        >>> # Returns DataFrame with one row
    """
    # Remove rows where required fields are missing
    required_cols = ['name', 'popularity', 'followers']
    df = df.dropna(subset=required_cols)
    
    # Standardize column names to snake_case
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    
    return df
```

# 4. Git Workflow Best Practices

<span style="display: inline-block; padding: 0 10px; font-size: 12px; font-weight: bold; line-height: 18px; white-space: nowrap; border: 1px solid #324655; border-radius: 2em; color: #fff; background-color: #324655; vertical-align: middle; box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);">GIT WORKFLOW</span>

Let's establish some clear guidelines for using Git effectively in your team project.

🔗 **Read more about [GitFlow](https://docs.github.com/en/get-started/using-github/github-flow).**

## 4.1 Issues: Your Task Management System

Issues help track work that needs to be done. Create them for:
- Features to implement (e.g., "Add minimal code to test the Spotify API")
- Bugs to fix
- Documentation to write
- Questions to discuss

<span style="display:block;background-color:rgba(93, 158, 188, 0.1);padding:0.5em;font-size:0.85em;margin-left:0em;margin-bottom:1em;border-radius:0.5em;width:50%">💡 <strong>Pro tip:</strong> Link issues to PRs using keywords like "Fixes #12" or "Resolves #15" in your PR description. GitHub will automatically close the issue when the PR is merged!</span>

🔗 [**GitHub documentation about Issues**](https://docs.github.com/en/issues)

## 4.2 Project Board: Your Visual Task Tracker

Think of it as your team's shared to-do list. Keep it simple:

| Column | Purpose | Example |
|--------|---------|---------|
| To Do | Planned work | "Set up database schema" |
| In Progress | Active work | "Collecting API data" |
| Review | Needs team check | "PR: Add visualization" |
| Done | Completed work | "Basic API setup" |

## 4.3 Creating and Using Branches

A branch represents a specific piece of work. Think of it as a container for related changes.


✅ **Do This:**

- **Create one branch per GitHub Issue** (add a new function, outline a notebook, fix a problem everyone's facing)
- Use clear naming conventions:

    ```output
    feature/add-spotify-api
    fix/authentication-error
    docs/update-readme
    ```
- Complete multiple commits on your branch until you've finished that feature

❌ **Don't Do This:**
- Create a new branch for every small change
- Use vague names like `johns-branch` or `test`
- Create multiple PRs from the same branch

## 4.4 Pull Requests: Your Code Review Tool

Pull Requests (PRs) are for reviewing completed features, not work in progress.

✅ **Good PR Practice:**

A good PR has a clear title and description, lists what it implements, and references the issue/task it resolves.

```markdown
Title: Add minimal code to test the Spotify API

This PR implements:
- Basic API authentication
- Data collection from 3 endpoints
- Rate limiting handling

Resolves #12

For the future: we will still need to check it returns data that is useful for our analysis.

```

❌ **Poor PR Practice:**
```markdown
Title: Update

Changed some stuff in the notebook
```
